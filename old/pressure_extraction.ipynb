```python
# Import modules
import paramiko
import os
from dataclasses import dataclass, asdict
import argparse
import json
from datetime import datetime
from pathlib import Path
from typing import Any, Optional
from stat import S_ISDIR, S_ISREG
import pandas as pd
from abc import ABC, abstractmethod
from lxml import etree
import matplotlib.pyplot as plt
```

```python
directory = 'D:\Daten\AVL_FIRE_ModOpt-PtX'
file_list = ['relative_pressure_reference_GHSV-211.dat', 'relative_pressure_GHSV-4000.dat', 'relative_pressure_GHSV-12000.dat', 
             'relative_pressure_GHSV-20000.dat', 'relative_pressure_GHSV-28000.dat', 'relative_pressure_GHSV-36000.dat']
data_frames = [pd.read_csv(os.path.join(directory, file), delim_whitespace=True, skiprows=[1]) for file in file_list]
```

```python
pressure_diffs = [(max(df['RelativePressure']) - min(df['RelativePressure'])) / 1e5 for df in data_frames]

x_values = [int(file_name.split('-')[-1].split('.')[0]) for file_name in file_list]
x_values
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(x_values, pressure_diffs, marker='o')
ax.set_xlabel('GHSV (1/h)') 
ax.set_ylabel('Pressure Difference (bar)')
ax.set_title('Pressure Difference vs GHSV')
```
